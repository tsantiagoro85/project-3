```python
# Import dependecies 
import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression
from scipy import stats
import scipy as sp
import matplotlib.pyplot as plt
import numpy as np
from string import ascii_letters
```

```python
# Read first dataset and preview
df = pd.read_csv("merged_datasets/dataset_income_2018-2023_melt.csv")
df.head()
```

```python
# Subset for Texas
df_Texas = df[(df["State"] == "Texas")]
df_Texas.head()
```

```python
# Subset for Total_Households categories 
df_Texas_Households = df_Texas.loc[df_Texas['Category'].str.startswith('Percent_Households')]
df_Texas_Households.head(10)
```

```python
# Plot and customize plots
plot = sns.barplot(df_Texas_Households, x = "Category", y = "Value", hue = "Year")

plot.set(xlabel = '', ylabel = 'Population %') # Add y axis label

plt.xticks(rotation = 45, horizontalalignment = 'right') # Rotate x axis label
plt.savefig("outputs/income_Texas.png", dpi = 300, bbox_inches = "tight") # Save figure
```
